```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Load your data into a DataFrame
movies_cleaned_df = pd.read_csv('tmdb_5000_movies.csv')

# Check if the DataFrame has been loaded correctly
print(movies_cleaned_df.head())
```

```python

# Check if the 'overview' column exists in the DataFrame
if 'overview' in movies_cleaned_df.columns:
    # Fill NaN values in 'overview' column with empty strings
    movies_cleaned_df['overview'] = movies_cleaned_df['overview'].fillna('')

    # Create an instance of TfidfVectorizer
    tfv = TfidfVectorizer()

    # Fit and transform the 'overview' column
    tfv_matrix = tfv.fit_transform(movies_cleaned_df['overview'])

    # Print the resulting TF-IDF matrix
    print(tfv_matrix)

    # Print the shape of the matrix
    print(tfv_matrix.shape)
else:
    print("The 'overview' column is not found in the DataFrame.")
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
print(sig[0])
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_cleaned.index, index=movies_cleaned['original_title']).drop_duplicates()
print(indices)
print(indices['Newlyweds'])
print(sig[4799])
print(list(enumerate(sig[indices['Newlyweds']])))
print(sorted(list(enumerate(sig[indices['Newlyweds']])), key=lambda x: x[1], reverse=True))
```

```python
def give_recomendations(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movies_cleaned['original_title'].iloc[movie_indices]
```

```python
print(give_recomendations('Avatar'))
```
